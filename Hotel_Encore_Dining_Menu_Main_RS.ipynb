```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://www.encorebostonharbor.com/menus/rare-steakhouse/main').text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
RS_Main_Dishes = []
for i in soup.find_all('span')[3:-2:2]:
    x = i.text.strip()
    RS_Main_Dishes.append(x)
    if(x=='Surf & Turf*'):
        break
```

```python
RS_Main_Price = []
for i in soup.find_all('span')[4:-1:2]:
    x = i.text.strip()    
    RS_Main_Price.append(x)
    if(x=='110'):
        break
```

```python
print(len(RS_Main_Dishes))
```

```python
print(len(RS_Main_Price))
```

```python
a = "traditional trimmings, brioche toast, blinis"
b = "Our aged steaks and chops are procured exclusively from the finest Midwestern ranchers."
c = "Wagyu cattle are known worldwide for producing meat with exceptional marbling characteristics and naturally enhanced flavor, tenderness, and juiciness. All of our Japanese Wagyu has a quality score of A5 and marble score between 8-10. Encore Boston Harbor is proud to be the only Kobe Beef Federation end user registered in Massachusetts."
```

```python
RS_Main_Description = []
for i in soup.find_all('p')[:-1]:
    x = i.text.strip()
    if(x!=a and x!=b and x!=c):
        RS_Main_Description.append(x)
    if(x=='4 oz. filet & half roasted Maine lobster'):
        break       
```

```python
print(len(RS_Main_Description))
```

```python
for i in soup.find_all('span')[69:-2:2]:
    x = i.text.strip()
    RS_Main_Dishes.append(x)
print(len(RS_Main_Dishes))    
```

```python
for i in soup.find_all('span')[70:-1:2]:
    x = i.text.strip()    
    RS_Main_Price.append(x)
RS_Main_Price
print(len(RS_Main_Price))
```

```python
RS_Main_Description
for i in soup.find_all('p')[36:-1]:
    x = i.text.strip()
    RS_Main_Description.append(x)
RS_Main_Description.remove("Wagyu cattle are known worldwide for producing meat with exceptional\nmarbling characteristics and naturally enhanced flavor, tenderness, and juiciness.\nAll of our Japanese Wagyu has a quality score of A5 and marble score between 8-10.\nEncore Boston Harbor is proud to be the only Kobe Beef Federation end user registered in Massachusetts.")
RS_Main_Description.remove("")
del RS_Main_Description[-3]
```

```python
len(RS_Main_Description)
```

```python
RS_Main_Menu = list(zip(RS_Main_Dishes , RS_Main_Description , RS_Main_Price ))
```

```python
RS_Main_Menu
```

```python
RS_Main_Menu_list = [[item for item in tpl] for tpl in RS_Main_Menu]
```

```python
RS_Main_Menu_list
```

```python
for i in range(0,57):
    if (i==0):
        RS_Main_Menu_list[i].append("CAVIAR SERVICE*")
    elif (i>=1 and i<7):
        RS_Main_Menu_list[i].append("CHILLED SEAFOOD & RAW BAR")
    elif (i>=7 and i<19):
        RS_Main_Menu_list[i].append("STARTERS, SOUPS & SALADS")
    elif (i>=19 and i<26):
        RS_Main_Menu_list[i].append("MAIN COURSES")
    elif (i>=26 and i<33):
        RS_Main_Menu_list[i].append("CHARRED PRIME CUT STEAKS & CHOPS")
    elif (i==33):
        RS_Main_Menu_list[i].append("CHARRED PRIME CUT STEAKS & CHOPS : DOUBLE CUT")
    elif (i>33 and i<38):
        RS_Main_Menu_list[i].append("CHARRED PRIME CUT STEAKS & CHOPS : SAUCE")
    elif (i>=38 and i<42):
        RS_Main_Menu_list[i].append("CHARRED PRIME CUT STEAKS & CHOPS : ADD TO THE STEAK")
    elif (i>=42 and i<54):
        RS_Main_Menu_list[i].append("SIDE DISHES")
    elif (i==54):
        RS_Main_Menu_list[i].append("JAPANESE WAGYU : KOBE, HYOGO PREFECTURE")
    elif (i==55):
        RS_Main_Menu_list[i].append("JAPANESE WAGYU : KUSUNOKI FARM, KAGOSHIMA PREFECTURE")
    elif (i==56):
        RS_Main_Menu_list[i].append("DOMESTIC WAGYU")
```
